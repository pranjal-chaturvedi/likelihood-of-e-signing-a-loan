# Feature engineering and one-hot encoding

```python
df["months_employed"] = df["months_employed"] + (df["years_employed"] * 12)
```

```python
df = df.drop("years_employed", axis=1)
```

```python
df.head()
```

```python
df["personal_account_m"] = df["personal_account_m"] + (df["personal_account_y"] * 12)
```

```python
df = df.drop("personal_account_y", axis=1)
```

```python
df.head()
```

```python
user_id = df["entry_id"]
```

```python
df = df.drop("entry_id", axis=1)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

## Train/Test split and preprocessing

```python
X = df.drop("e_signed", axis=1)
```

```python
y = df["e_signed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Training the model - Random Forest Classifier with Grid Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
base_model = RandomForestClassifier()
```

```python
param_grid = {"criterion": ["gini", "entropy"],
              "max_depth": [3, 4, None],
              "min_samples_split": [2, 4, 7, 10],
              "bootstrap": [True, False]}
```

```python
model = GridSearchCV(estimator=base_model, param_grid=param_grid, cv=10, scoring="accuracy")
```

```python
model.fit(scaled_X_train, y_train)
```

```python
model.best_params_
```

```python
predictions = model.predict(scaled_X_test)
```

### Evaluating the model

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
plot_confusion_matrix(model, scaled_X_test, y_test)
```

```python
print(classification_report(y_test, predictions))
```
